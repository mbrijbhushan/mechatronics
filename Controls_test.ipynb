```python
from control import * #Load Python Control Systems Library
import matplotlib.pyplot as plt #Plotting library
import numpy as np #Numerical library
```

```python
s=tf([1,0],[1]) #Define 's' variable
s
```

```python
sys1=10000*((s+20)*(s+10))/((s+100)*(s*(s**2+20*s+400))) #Define transfer function
sys1
```

```python
pzmap(sys1);
```

```python
#Root locus
root_locus(sys1);
```

```python
bode_plot(sys1); #Plot the frequency response bode plot
```

```python
#Closed loop
cl_sys1 = feedback(sys1,1)
cl_sys1
```

```python
#Plot the step response
T1 = np.arange(0,0.5,0.001)
T, yout = step_response(cl_sys1,T1)
plt.plot(T,yout)
plt.title('Step Response')
plt.xlabel('Time')
plt.ylabel('Position');
```

```python
#Nyquist Plot
nyquist_plot(sys1);
```

```python
# Gain margin and phase margin
margin(sys1)
```
